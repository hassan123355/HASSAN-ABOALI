```python
# 1. Print even numbers between 1 and 100
for i in range(1, 101):
    if i % 2 == 0:
        print(i)
```

```python


# 2. Sum of digits
num = int(input("Enter number: "))
s = 0
while num > 0:
    s += num % 10
    num //= 10
print("Sum =", s)
```

```python


# 3. Factorial
n = int(input("Enter number: "))
fact = 1
for i in range(1, n+1):
    fact *= i
print("Factorial =", fact)
```

```python

# 4. Multiplication table

n = int(input("Enter number: "))
for i in range(1, 11):
    print(n, "x", i, "=", n*i)
```

```python

# 5. Prime number
n = int(input("Enter number: "))
prime = True
for i in range(2, n):
    if n % i == 0:
        prime = False
        break

if prime and n > 1:
    print("Prime")
else:
    print("Not Prime")
```

```python

# 6. Reverse integer
num = int(input("Enter number: "))
rev = 0
while num > 0:
    rev = rev*10 + num%10
    num //= 10
print("Reverse =", rev)
```

```python

# 7. Count vowels
text = input("Enter string: ")
count = 0
for ch in text:
    if ch.lower() in "aeiou":
        count += 1
print("Vowels =", count)
```

```python

# 8. Guess random number
import random
num = random.randint(1, 20)

guess = 0
while guess != num:
    guess = int(input("Guess: "))
print("Correct!")
```

```python

# 9. Largest of three
def largest(a, b, c):
    return max(a, b, c)
```

```python

# 10. Fibonacci
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a+b
```

```python

# 11. Palindrome string
def is_pal(s):
    return s == s[::-1]
```

```python
# 12. Count even & odd in list
def count_even_odd(lst):
    even = odd = 0
    for n in lst:
        if n % 2 == 0:
            even += 1
        else:
            odd += 1
    return even, odd
```

```python
# 13. Grade function
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
```
